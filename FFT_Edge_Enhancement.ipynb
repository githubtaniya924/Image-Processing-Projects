# FFT Edge Enhancement

```python
from skimage import data
import numpy as np
import matplotlib.pyplot as plt

img_volume = data.brain()
img = img_volume[0]

img_f = img.astype(np.float32)

F = np.fft.fft2(img_f)
F_shift = np.fft.fftshift(F)

mag = np.log(1+np.abs(F_shift))

rows,cols = img.shape
crows,ccols = rows//2,cols//2

mask = np.ones((rows,cols),np.uint8)
mask[crows-20:crows+20,ccols-20:ccols+20] = 0

F_hp = F_shift * mask

F_ishift = np.fft.ifftshift(F_hp)
img_edges = np.abs(np.fft.ifft2(F_ishift))

plt.figure(figsize=(14,8))

plt.subplot(1,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original MRI Slice')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mag,cmap='gray')
plt.title('FFT Magnitude Spectrum')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(img_edges,cmap='gray')
plt.title('After High-Pass Filtering (Edges)')
plt.axis('off')

plt.tight_layout()
plt.show()

```
