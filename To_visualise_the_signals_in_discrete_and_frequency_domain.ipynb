# **To visualise the signals in discrete and frequency domain. **

```python
import numpy as np #array, sin, FFT
import matplotlib.pyplot as plt #for plotting graphs

n= np.arange(0,64)
#create a discrete-time signal
#example signal: combination of two sinusoids
#mathematically: x[n] = 2*sin(0.2*pi*n) + sin(0.6*pi*n)
x = 2*np.sin(0.2*np.pi*n)+ np.sin(0.6*np.pi*n)

X =np.fft.fft(x) #compute FFT

freq = np.fft.fftfreq(len(x)) #frequency bins

X_mag = np.abs(X) #Magnitude Spectrum

X_Phase = np.angle(X) #Phase Spectrum
```

```python
plt.figure(figsize=(14,10))

#time domain
plt.subplot(3,1,1)
plt.stem(n,x,basefmt=" ")
plt.title("Discrete-Time Signal (Time Domain)")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)
```

```python
#frequency vs Magnitude
#magnitude Spectrum
plt.subplot(3,1,2)
plt.stem(freq,X_mag,basefmt=" ")
plt.title("Magnitude Spectrum[X][k]")
plt.xlabel("Normalized Frequency")
plt.ylabel("Magnitude")
plt.grid(True)
```

```python
#Phase Spectrum
plt.subplot(3,1,3)
plt.stem(freq,X_Phase,basefmt=" ")
plt.title("Phase Spectrum[X][k] (Frequency Domain)")
plt.xlabel("Normalized Frequency")
plt.ylabel("Phase (radians)")
plt.grid(True)
```
