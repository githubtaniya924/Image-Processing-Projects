```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt

#upload an image
uploaded = files.upload()

#get the uploaded file name
image_path = list(uploaded.keys())[0]

#read the image
image = cv2.imread(image_path,cv2.IMREAD_GRAYSCALE)

#check if image is loaded
if image is None:
  raise ValueError("Error loading images. please upload a valid image file")

#Normaliza image to [0,1]
img_norm = image.astype(np.float32)/255.0

relu_image = np.maximum(0,img_norm)

leaky_relu_image = np.where(img_norm>0,img_norm,0.1*img_norm)

sigmoid_image = 1/(1+np.exp(-img_norm*5))  #contrast factor = 5

tanh_image = np.tanh(img_norm*3) #scaling factor = 3

#step function (threshold at 0.5)
step_image = np.where(img_norm>=0.5,1,0)

titles = [
          "Original Image",
          "ReLU",
          "Leaky ReLU",
          "Sigmoid",
          "Tanh",
          "Step Function Output"
]

images =[
    img_norm,
    relu_image,
    leaky_relu_image,
    sigmoid_image,
    tanh_image,
    step_image
]

plt.figure(figsize=(12,10))

for i in range(6):
  plt.subplot(2,3,i+1)
  plt.imshow(images[i],cmap='gray')
  plt.title(titles[i])
  plt.axis('off')

plt.tight_layout()
plt.show()
```
